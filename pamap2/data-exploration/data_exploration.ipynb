```python
def rolling_frame(df, length, hop):
    # one row equals 10ms
    start = 0
    end = len(df)

    frame_begin = start
    frame_end = start + length
    samples = []
    while frame_end < end:
        df1 = df[frame_begin:frame_end]
        samples.append(df1)
        frame_begin = frame_begin + hop
        frame_end = frame_end + hop
    return samples


print(len(rolling_frame(data_per_activity['1'], 300, 16)))
```
